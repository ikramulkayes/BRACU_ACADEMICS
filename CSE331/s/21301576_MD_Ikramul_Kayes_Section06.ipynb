```python
import itertools


X = {"S": "AB,BC",
    "A":"BA,a",
    "B":"CC,b",
    "C":"AB,a"}

y = "baaba"

def CYK_Algo(X,y):   #dictionary for variable and terminal rules
    dic_store = {}
    for key,val in X.items():
        for p in val.split(","):
            if p not in dic_store.keys():
                dic_store[p] = key
            else:
                dic_store[p] += key
    #print(dic_store)
    
    #Making the 2d dynamic list 
    
    lst_store = []
    length = len(y)
    for elem in range(length-1,-1,-1):
        lst_store.append([])
        for i in range(elem+1):
            lst_store[-1].append([])
    
    

    #Changing the first index of dynamic array using the given string
    count = 0
    for s in y:
        d = dic_store[s]
        for vals in d:
            lst_store[0][count].append(vals)
        count += 1
    #print(lst_store)

    #working the rest of the dynamic array 

    for q in range(1,length):
        for val1 in range(length-q):
            t_lst_store = []
            for val2 in range(q):       #stair up & down
                idx1 = lst_store[val2][val1]
                idx2 = lst_store[q-val2-1][val1+val2+1]
                idx3 = list(itertools.product(idx1,idx2))
                #print(idx3)
                t_lst_store.append(idx3)  #storing calculative results 
            
            new_idx = []               
            for ss in t_lst_store:      #accessing values from stored calculative result list      
                for u in ss:
                    if len(u) ==0:     #if the tuple is empty, continue
                        continue
                    u1,u2 = u
                    
                    if len(u1)==0 or len(u2)==0:  #if any of the these are empty,avoid it
                        pass
                    else:
                        uk = u1 + u2              #merging these      
                        #print(uk)
                        if uk in dic_store.keys():    
                            ug = dic_store[uk]
                            #print(ug)
                            
                            new_idx.append(ug)
                            
            for b in new_idx:   
                for cc in b:             #checking duplicate variables
                    if cc not in lst_store[q][val1]:  #excluding duplicate variables
                        lst_store[q][val1].append(cc)
    
    check= lst_store[-1]  
    if 'S' in check[0]:                 #printing result
        print("YES, possible!")
    else:
        print("NO, not possible!")
        
    print(lst_store)


g = CYK_Algo(X,y)
```
